```python
from transformers.models.t5 import T5Config, T5ForConditionalGeneration
from transformers import T5Tokenizer, BertTokenizer, BertModel
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import spacy

nlp = spacy.load("en_core_web_sm")

import os
os.chdir("/Users/jaceysimpson/Vscode/personal-manager-02")
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')
bert_model.eval()
data = pd.read_csv('data/event_scheduling.csv')
test_data = data[:5]

texts = test_data['input_text']
for i, text in enumerate(texts):
    encoded_inputs = tokenizer(
        text,
        return_tensors='pt',
        padding='max_length',
        truncation=True,
        max_length=100)
    token_ids = encoded_inputs['input_ids']
    attention_mask = encoded_inputs['attention_mask']
    string_text = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(token_ids[0]))
    doc = nlp(string_text)
    tags = [token.pos_ for token in doc]

    with torch.no_grad():
        outputs = bert_model(token_ids, attention_mask=attention_mask)
        hidden_states = outputs.last_hidden_state
    print(f"Original Text {i}: {string_text}\n Tags: {tags}\n Token ID Shape: {token_ids.shape} \n Hidden Shape: {hidden_states.shape}\n Token IDs: {token_ids}\n Attention Mask: {attention_mask}\n")
```
